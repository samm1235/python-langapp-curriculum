Before we get started, we need to clean our data. There are acutally some duplicated words in our file! Verify for yourself that there are indeed duplicated words.



```python
with open("data/words.txt","r") as files:
    lines = files.readlines()
    for index, word in enumerate(lines):
        word = word.strip()
        lines[index] = word

print(lines)
```
